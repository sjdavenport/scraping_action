```python
import bilbystats as bs
import json
import pandas as pd
from datetime import datetime
import re

html_path = "../data/SNP500/23-01-2026.html"

with open(html_path, "r", encoding="utf-8") as f:
    html = f.read()
```

```python
bs.show_html(html)
```

```python
pattern = re.compile(
    r'TSDtV":"(%\.@.\[.*?\])"',
    re.DOTALL
)

match = pattern.search(html)
if not match:
    raise ValueError("Could not find intraday price data in HTML")

raw_data = match.group(1)
```

```python
raw_data
```

```python
cleaned = (
    raw_data
    .replace('\\u003d', '=')
    .replace('\\u0026', '&')
    .replace('\\"', '"')
)

# Remove Google-specific prefix
cleaned = cleaned.replace('%.@.', '')
```

```python
import json

# Remove Google-specific prefix
cleaned = cleaned.replace('%.@.', '').strip()

# --- NEW PART ---
# Find the first complete top-level JSON array
bracket_count = 0
end_idx = None

for i, ch in enumerate(cleaned):
    if ch == '[':
        bracket_count += 1
    elif ch == ']':
        bracket_count -= 1
        if bracket_count == 0:
            end_idx = i + 1
            break

if end_idx is None:
    raise ValueError("Could not isolate JSON array")

json_array_str = cleaned[:end_idx]

data = json.loads(json_array_str)
```
