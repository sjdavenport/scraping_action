```python
import pandas as pd
import bilbystats as bs
```

```python
df = pd.read_excel('/Users/samd/Documents/Projects/scraping/financial_scraping/notebooks/ftse250.xlsx', engine = 'calamine')
```

```python
df
```

```python
starter = "https://www.google.com/finance/quote/"
ender = ':LON?hl=en'
tickers = df['tickers'].to_list()
all_links = []
for i in range(len(df)):
    all_links.append(starter + tickers[i] + ender)
```

```python
all_links
```

```python
import json
result = []

for i in range(len(all_links)):
    name = tickers[i]
    url = all_links[i]
    result.append({"name": name, "url": url})

# Write to JSON file
with open('stocks.json', 'w') as f:
    json.dump(result, f, indent=2)
```
